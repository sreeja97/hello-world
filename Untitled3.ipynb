```
import glob
import cv2
import numpy as np
import matplotlib.pyplot as plt
class_names = ['mucca','pecora']
IMG_SIZE = 224
```

```
def get_data_from_folder(path):
  data = []
  for class_name in class_names: # iterate through classes
    try:
      files = glob.glob(path+"/*") # get files in each folder(class)
      for f in files:
        img = cv2.imread(f) #read the image
        img = cv2.resize(img,(IMG_SIZE,IMG_SIZE)) #resize the image
        gray_img = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
        data.append([np.array(gray_img),class_names.index(class_name)]) #Append images and corresponding labels to data
    except:
      pass
  np.random.shuffle(data)
  return data
```
