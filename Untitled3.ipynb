```
import glob
import cv2
import numpy as np
import matplotlib.pyplot as plt
class_names = ['mucca','pecora']
IMG_SIZE = 224
```

```
def get_data_from_folder(path):
  x_train=[]
  y_train=[]
  for class_name in class_names: # iterate through classes
    try:
      files = glob.glob(path+"/*") # get files in each folder(class)
      for f in files:
        img = cv2.imread(f) #read the image
        img = cv2.resize(img,(IMG_SIZE,IMG_SIZE)) #resize the image
        gray_img = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
        x_train.append(np.array(gray_img))
        y_train.append(class_names.index(class_name))
    except:
      pass
  np.random.shuffle(x_train)
  np.random.shuffle(y_train)
  return x_train, y_train
```

```
training_data_x, training_data_y = get_data_from_folder('/content/train_beg')

training_data_y[0]
```

```
import pandas as pd
labels_csv = pd.read_csv('https://raw.githubusercontent.com/dphi-official/Datasets/master/animal_data/Training_set_animals.csv')
labels_csv.head()   # this file contains file name and corresponding labels
labels_csv.loc[(labels_csv['filename'] == 'Image_1.jpg') & (labels_csv['animal_type'] == 'mucca')]
```

```
import tensorflow as tf
import keras
import sklearn as sk
from keras.datasets import mnist
from keras.models import Sequential
from keras.layers import Dense, Dropout, Conv2D, MaxPool2D
from keras.utils import np_utils
from sklearn.model_selection import train_test_split
X_train, X_test, Y_train, Y_test = train_test_split(training_data_x, training_data_y, random_state=42, test_size=0.2)
```

```
X_train[0]
```

```
Y_train[0]
```

```
X_train=tf.convert_to_tensor(
    X_train, dtype='float32', dtype_hint=None, name=None
)

X_train.get_shape
```

```
Y_train = tf.convert_to_tensor(
    Y_train, dtype=None, dtype_hint=None, name=None
)
Y_train.shape
```

```
X_test = tf.convert_to_tensor(
    X_test, dtype='float32', dtype_hint=None, name=None
)
X_test.shape
```

```
Y_test = tf.convert_to_tensor(
    Y_test, dtype=None, dtype_hint=None, name=None
)
Y_test.shape
```

```
 X_train_np =np.asarray(X_train)
 Y_train_np = np.asarray(Y_train)
 X_test_np = np.asarray(X_test)
 Y_test_np = np.asarray(Y_test)
```

```
X_train_np.shape
```

```
X_train_org = X_train_np.reshape(X_train_np.shape[0], 224, 224, 1)
X_test_org = X_test_np.reshape(X_test_np.shape[0], 224, 224, 1)
```

```
X_train_org = X_train_org.astype('float32')
X_test_org = X_test_org.astype('float32')
```

```
X_train_org[0]
```

```
X_train_org /= 255
X_test_org /= 255
```

```
X_train_org[0]
```

```
from keras.models import Sequential
from keras.layers import Dense, Dropout, Conv2D, MaxPool2D, Flatten
from keras.utils import np_utils
```

```
from keras.optimizers import SGD, RMSprop
model = Sequential()
model.add(Conv2D(16, (3, 3), activation='relu', input_shape=(224, 224, 1)))
model.add(MaxPool2D((2, 2)))
model.add(Conv2D(32, (3, 3), activation='relu'))
model.add(MaxPool2D((2, 2)))
model.add(Conv2D(64, (3, 3), activation='relu'))
model.add(MaxPool2D((2, 2)))
model.add(Flatten())
model.add(Dense(512, activation='relu'))
model.add(Dense(1, activation='sigmoid'))
model.summary()
```

```
model.compile(loss='binary_crossentropy', optimizer='sgd', metrics=['accuracy'])

```

```
from keras.callbacks import EarlyStopping

# Define a callback to monitor val_acc
monitor_val_acc = EarlyStopping(monitor="val_loss", 
                       patience=5)

# Train your model using the early stopping callback
model.fit(X_train_org, Y_train_np, steps_per_epoch=8,verbose=1,
           epochs=15, validation_data=(X_test_org,Y_test_np),
           callbacks=[monitor_val_acc],
          )
```

```
from sklearn import metrics
y_pred = model.predict(X_test_org)
y_pred = (y_pred >= 0.65)*1
cnf_matrix = metrics.confusion_matrix(Y_test_np, y_pred)
cnf_matrix
```

```
print("Accuracy:",metrics.accuracy_score(Y_test_np, y_pred))
print("Precision:",metrics.precision_score(Y_test_np, y_pred))
print("Recall:",metrics.recall_score(Y_test_np, y_pred))
```
